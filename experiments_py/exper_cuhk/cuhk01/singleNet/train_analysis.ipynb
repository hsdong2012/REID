```python
fp=open('./1130log.txt','r')
f_lines = fp.readlines()
# print f_lines
fp.close()
iter_list = []
accuracy_list = []
loss_list = []
for line in f_lines:
    if line.find('Testing net (#0)')!=-1:
#         print line
        end = line.find('Testing net (#0)')
        iter_num = line[end-15:end-2]
        iter_list.append(int(iter_num.split(' ')[-1]))
#         print iter_num.split(' ')[-1]
    if line.find('Test net output #0: accuracy =')!=-1:
        begin= line.find('Test net output #0: accuracy =')+31
        accuracy = line[begin:begin+8]
        accuracy_list.append(float(accuracy.split(' ')[0]))
#         print accuracy
#         print line
    if line.find('Test net output #1: loss =')!=-1:
        begin= line.find('Test net output #1: loss =')+27
        loss = line[begin:begin+8]
#         print loss
        loss_list.append(float(loss.split(' ')[0]))
#         print accuracy
# print loss_list
```

```python
import matplotlib.pyplot as plt
get_ipython().magic(u'matplotlib inline')   
from matplotlib.legend_handler import HandlerLine2D
import numpy as np
plt.ioff()

fig = plt.figure(figsize=(6,5),dpi=180)

# x_plot=[temp+1 for temp in x_line]

# plt.plot(x_plot, accuracy_list[x_line],'r-',label='accuracy')
plt.plot(iter_list,accuracy_list,'r-',label='accuracy')

#start plot
# plt.plot(x_plot, loss_list[x_line],'b-',label='loss')
plt.plot(iter_list,loss_list,'b-',label='loss')
plt.xlabel('iter_num')
plt.ylabel('loss or accuracy')
plt.grid()
plt.legend(loc=4)
# plt.savefig(pathname)
plt.show()
```
